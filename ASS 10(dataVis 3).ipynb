```python
import pandas as pd
import seaborn as sns
```

```python
dataset = sns.load_dataset("iris")
dataset.head()   
# we have 5 features
# first 4 are numerical
#1) LISTING DOWN
# last is categorical,nominal feature
```

```python
#  1) CREATING histogram for each feature
```

```python
import matplotlib.pyplot as plt
fig,axes =plt.subplots(2,2,figsize=(16,9)) # displaying grid 2 X 2 grid 
sns.histplot(dataset["sepal_length"], ax=axes[0,0])# plot histogram | axes specifies which quadrant to display in
sns.histplot(dataset["sepal_width"], ax=axes[0,1])
sns.histplot(dataset["petal_length"],ax=axes[1,0])
sns.histplot(dataset["petal_width"])       # this will by default go to last quadrant


```

```python
# 1) INFERNCE/OBSERVATION

# sepal_lenth is evenly distributed for COUNT from 0 -25+

#sepal_width is normal distibution for COUNT  from 0 -35+

# petal_length and petal_width are left skewd as value on  left is more
```

```python
# 2)

import matplotlib.pyplot as plt

fig,axes =plt.subplots(2,2,figsize=(16,9))

sns.boxplot(y="petal_length", x="species", data=dataset,ax=axes[0,0])
sns.boxplot(y="petal_width", x="species", data=dataset,ax=axes[0,1])
sns.boxplot(y="petal_length",x="species",data=dataset,ax=axes[1,0])
sns.boxplot(y="petal_width",x="species",data=dataset,ax=axes[1,1])
```
